```python
import pandas as pd #do pracy z danymi w tabelkach
import matplotlib.pyplot as mp #wykresy
import seaborn as sb #bardzije zaawansowane wykresy
import numpy as np #do pracy z macierzami
import datetime #do pracy z datami
```

```python
#dane z Meteo

filename = input("Podaj nazwę pliku do wczytani: ")
meteo = pd.read_csv(f"../data/processed/meteo/{filename}")
meteo = meteo.dropna() #usuwamy wiersze z brakującymi danymi
```

```python
meteo.info()
meteo.head(20)
#meteo.describe()
```

```python
import matplotlib.dates as mdates
meteo.columns
subset = meteo.head(30) #wybieramy pierwsze 50 wierszy
```

```python
mp.figure(figsize=(12, 6))
mp.title("Parametry powietrza w czasie (godzinowym)")
mp.xlabel("Czas")
mp.ylabel("Wartość")


# Formatowanie osi X – co 4 godziny
mp.gca().xaxis.set_major_locator(mdates.HourLocator(interval=24))
mp.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M%d-%m'))


mp.plot(subset["time"], subset["temperature_2m"], label="temperature")
mp.plot(subset["time"], subset["relative_humidity_2m"], label="humidity")
mp.plot(subset["time"], subset["windspeed_10m"], label="windspeed")
mp.legend() 
mp.xticks(rotation=45)
mp.show()
```
